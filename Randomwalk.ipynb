```python
import numpy as np 
import matplotlib.pyplot as plt 
  

prob = [0.1, 0.7]   
  

start = 0  
positions = [start] 
  

x = np.random.normal(1000) 
down = x < prob[0] 
up = x > prob[1] 
  
  
for idownp, iupp in zip(downp, upp): 
    down = idownp and positions[-1] > -1
    up = iupp and positions[-1] < 1
    positions.append(positions[-1] - down + up) 
  

plt.plot(positions) 
plt.show() 
```
